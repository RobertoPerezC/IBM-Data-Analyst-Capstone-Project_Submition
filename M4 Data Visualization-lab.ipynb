# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
#Import the libraries needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%pip install -U seaborn
```

```python
# Define the query
QUERY1 = """
SELECT ConvertedComp
FROM master
"""
# Execute the query and read the result into a DataFrame
df = pd.read_sql_query(QUERY1, conn)
# Plot histogram
df['ConvertedComp'].plot.hist(figsize=(10, 6))
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
# Close the connection to the database
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
# Define the query
QUERY2 = """
SELECT Age
FROM master
"""
# Execute the query and read the result into a DataFrame
df_age = pd.read_sql_query(QUERY2, conn)
# Plot box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', data=df_age)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# Define the query
QUERY3 = """
SELECT Age, WorkWeekHrs
FROM master
"""
# Execute the query and read the result into a DataFrame
df_age_workweek = pd.read_sql_query(QUERY3, conn)
# Plot scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df_age_workweek)
plt.title('Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Define the query
QUERY4 = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
# Execute the query and read the result into a DataFrame
df_age_work_rev = pd.read_sql_query(QUERY4, conn)
# Plot bubble plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20, 200), data=df_age_work_rev)
plt.title('WorkWeekHrs vs CodeRevHrs')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
# Define the query to get the top 5 databases respondents wish to learn next year
QUERY5 = """
SELECT DatabaseDesireNextYear as Database, COUNT(*) as No_of_respondents
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear 
ORDER BY No_of_respondents DESC LIMIT 5
"""
# Execute the query and read the result into a DataFrame
df_top_databases = pd.read_sql_query(QUERY5, conn)
df_top_databases
```

```python
# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_top_databases['No_of_respondents'], labels=df_top_databases['Database'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Define the query to get the median WorkWeekHrs and CodeRevHrs for age group 30 to 35
QUERY6 = """
SELECT Age, 
       CASE
           WHEN Age BETWEEN 30 AND 35 THEN '30-35'
           ELSE 'Other'
       END AS Age_Group,
       WorkWeekHrs,
       CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Execute the query and read the result into a DataFrame
df_age_group = pd.read_sql_query(QUERY6, conn)

# Calculate median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
median_data = df_age_group.groupby('Age_Group').median()

# Plot stacked chart
median_data[['WorkWeekHrs', 'CodeRevHrs']].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)
plt.legend(title='Hours')
plt.show()
```

```python
# Define the query to get the WorkWeekHrs and CodeRevHrs for each age between 30 and 35
QUERY6 = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Execute the query and read the result into a DataFrame
df_age_group = pd.read_sql_query(QUERY6, conn)

# Plot stacked chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='WorkWeekHrs', data=df_age_group, color='blue', label='WorkWeekHrs')
sns.barplot(x='Age', y='CodeRevHrs', data=df_age_group, color='orange', label='CodeRevHrs')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# Define the query to get the median ConvertedComp for ages between 45 and 60
QUERY7 = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the query and read the result into a DataFrame
df_age_comp = pd.read_sql_query(QUERY7, conn)

# Calculate median ConvertedComp for each age
median_comp = df_age_comp.groupby('Age').median()

# Plot line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=median_comp, x=median_comp.index, y='ConvertedComp')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.xticks(rotation=45)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Define the query to get the MainBranch column
QUERY8 = """
SELECT MainBranch
FROM master
"""

# Execute the query and read the result into a DataFrame
df_main_branch = pd.read_sql_query(QUERY8, conn)

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
sns.countplot(y='MainBranch', data=df_main_branch, order=df_main_branch['MainBranch'].value_counts().index)
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```
